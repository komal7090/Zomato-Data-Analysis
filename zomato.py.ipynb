```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\Ap\Downloads\Zomato\zomato.csv", encoding='latin-1')
df.head()
```

```python
df = df.drop(['Address'], axis=1)
df.head()
```

```python
# Check for missing (NaN) values in each column
df.isnull().sum()

```

```python
# Remove rows with missing values
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
# Basic info about data types and counts
df.info()

# Summary statistics for numeric columns
df.describe()
```

```python
top_cities = df['City'].value_counts().head(10)
print(top_cities)
```

```python
top_rated = df.sort_values(by='Aggregate rating', ascending=False).head(10)
top_rated[['Restaurant Name', 'City', 'Aggregate rating', 'Votes']]
```

```python
import matplotlib.pyplot as plt

# Top 10 cities with most restaurants
top_cities.plot(kind='bar', color='orange')
plt.title('Top 10 Cities with Most Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Count the number of restaurants for each cuisine
top_cuisines = df['Cuisines'].value_counts().head(10)
print(top_cuisines)

# Optional: Bar chart
top_cuisines.plot(kind='bar', color='green')
plt.title('Top 10 Most Popular Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Average cost by city
avg_cost = df.groupby('City')['Average Cost for two'].mean().sort_values(ascending=False).head(10)
print(avg_cost)

# Optional: Bar chart
avg_cost.plot(kind='bar', color='purple')
plt.title('Average Cost for Two by City (Top 10)')
plt.ylabel('Average Cost for Two')
plt.show()
```

```python
# Restaurants with high rating and votes
top_restaurants = df[df['Aggregate rating']>=4.5].sort_values(by='Votes', ascending=False).head(10)
top_restaurants[['Restaurant Name', 'City', 'Cuisines', 'Aggregate rating', 'Votes']]
```

```python
import seaborn as sns

# Distribution of ratings
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='blue')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Aggregate Rating')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

plt.figure(figsize=(18,12))

# 1️⃣ Top 10 Cities by Restaurants
plt.subplot(2,2,1)
top_cities.plot(kind='bar', color='orange')
plt.title('Top 10 Cities with Most Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')

# 2️⃣ Top 10 Cuisines
plt.subplot(2,2,2)
top_cuisines.plot(kind='bar', color='green')
plt.title('Top 10 Most Popular Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')

# 3️⃣ Average Cost by City
plt.subplot(2,2,3)
avg_cost.plot(kind='bar', color='purple')
plt.title('Average Cost for Two by City (Top 10)')
plt.ylabel('Average Cost for Two')

# 4️⃣ Rating Distribution
plt.subplot(2,2,4)
sns.histplot(df['Aggregate rating'], bins=20, kde=True, color='blue')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Aggregate Rating')

plt.tight_layout()
plt.show()
```
